```python
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
from PIL import Image, ImageTk
import test

class WelcomePage:
    def __init__(self, root):
        self.root = root
        self.root.title("Job Scheduler - Welcome")
        self.root.state('zoomed')
        
        # Set background color
        bg_color = "#E8F3E8"  # Light sage green
        self.root.configure(bg=bg_color)
        
        # Main frame with styling
        style = ttk.Style()
        style.configure('Welcome.TFrame', background=bg_color)
        style.configure('Title.TLabel', background=bg_color, font=('Jacquard', 32, 'bold'))
        style.configure('Subtitle.TLabel', background=bg_color, font=('Helvetica', 18))
        style.configure('Algorithm.TButton', font=('Helvetica', 12), padding=10)
        
        main_frame = ttk.Frame(self.root, padding="40", style='Welcome.TFrame')
        main_frame.place(relx=0.5, rely=0.5, anchor="center")
        
        # Title with decorative border
        title_frame = ttk.Frame(main_frame, style='Welcome.TFrame')
        title_frame.grid(row=0, column=0, pady=(0, 20))
        
        title_label = ttk.Label(title_frame, 
                              text="Job Scheduling Problem Solving",
                              style='Title.TLabel')
        title_label.grid(row=0, column=0, pady=10)
        
        # Subtitle
        subtitle_label = ttk.Label(main_frame, 
                                 text="You Want It When?",
                                 style='Subtitle.TLabel',
                                 foreground="red")
        subtitle_label.grid(row=1, column=0, pady=10)
        
        # Second subtitle
        choose_label = ttk.Label(main_frame,
                               text="Choose Your Algorithm",
                               style='Subtitle.TLabel')
        choose_label.grid(row=2, column=0, pady=(30, 20))
        
        # Buttons frame
        button_frame = ttk.Frame(main_frame, style='Welcome.TFrame')
        button_frame.grid(row=3, column=0, pady=30)
        
        # Algorithm buttons
        backtracking_btn = ttk.Button(button_frame, 
                                    text="Backtracking",
                                    style='Algorithm.TButton',
                                    command=lambda: self.start_scheduler("Backtracking"))
        backtracking_btn.grid(row=0, column=0, padx=20)
        
        genetic_btn = ttk.Button(button_frame,
                               text="Genetic Algorithm",
                               style='Algorithm.TButton',
                               command=lambda: self.start_scheduler("Genetic"))
        genetic_btn.grid(row=0, column=1, padx=20)
        
        # Configure grid weights for main_frame
        main_frame.grid_columnconfigure(0, weight=1)
        
        # Configure grid weights for button_frame
        button_frame.grid_columnconfigure((0, 1), weight=1)
        
        # Configure root grid weights
        self.root.grid_rowconfigure(0, weight=1)
        self.root.grid_columnconfigure(0, weight=1)
        
    def start_scheduler(self, algorithm):
        self.root.destroy()
        new_root = tk.Tk()
        app = SimpleSchedulerGUI(new_root, initial_algorithm=algorithm)
        new_root.mainloop()

class SimpleSchedulerGUI:
    def __init__(self, root, initial_algorithm="Backtracking"):
        self.root = root
        self.root.title("Simple Job Scheduler")
        self.root.state('zoomed')
        
        # Set background color
        bg_color = "#E8F3E8"  # Light sage green
        self.root.configure(bg=bg_color)
        
        # Style configuration
        style = ttk.Style()
        style.configure('Main.TFrame', background=bg_color)
        style.configure('Main.TLabelframe', background=bg_color)
        style.configure('Main.TLabelframe.Label', background=bg_color, font=('Helvetica', 10))
        style.configure('Main.TLabel', background=bg_color)
        style.configure('Main.TButton', font=('Helvetica', 10), padding=5)
        
        # Main frame
        main_frame = ttk.Frame(self.root, padding="10", style='Main.TFrame')
        main_frame.grid(row=0, column=0, sticky="nsew")
        
        # Left frame for basic inputs
        left_frame = ttk.LabelFrame(main_frame, text="Basic Data", padding="5", style='Main.TLabelframe')
        left_frame.grid(row=0, column=0, sticky="nsew", padx=5, pady=5)
        
        # Job inputs
        ttk.Label(left_frame, text="Number of Jobs:", style='Main.TLabel').grid(row=0, column=0, padx=5, pady=2)
        self.num_jobs = ttk.Entry(left_frame, width=10)
        self.num_jobs.grid(row=0, column=1, padx=5, pady=2)
        self.num_jobs.insert(0, "2")
        
        ttk.Label(left_frame, text="Job Durations (comma-separated):", style='Main.TLabel').grid(row=1, column=0, padx=5, pady=2)
        self.job_durations = ttk.Entry(left_frame, width=30)
        self.job_durations.grid(row=1, column=1, padx=5, pady=2)
        self.job_durations.insert(0, "2,3")
        
        # Machine inputs
        ttk.Label(left_frame, text="Number of Machines:", style='Main.TLabel').grid(row=2, column=0, padx=5, pady=2)
        self.num_machines = ttk.Entry(left_frame, width=10)
        self.num_machines.grid(row=2, column=1, padx=5, pady=2)
        self.num_machines.insert(0, "2")
        
        ttk.Label(left_frame, text="Machine Capacities (comma-separated):", style='Main.TLabel').grid(row=3, column=0, padx=5, pady=2)
        self.machine_caps = ttk.Entry(left_frame, width=30)
        self.machine_caps.grid(row=3, column=1, padx=5, pady=2)
        self.machine_caps.insert(0, "1,1")
        
        # Algorithm selection
        ttk.Label(left_frame, text="Algorithm:", style='Main.TLabel').grid(row=4, column=0, padx=5, pady=2)
        self.algorithm = ttk.Combobox(left_frame, values=["Backtracking", "Genetic"], state="readonly")
        self.algorithm.grid(row=4, column=1, padx=5, pady=2)
        self.algorithm.set(initial_algorithm)
        
        # Auto-split option
        self.auto_split = tk.BooleanVar(value=True)
        ttk.Checkbutton(left_frame, text="Auto-assign jobs to machines", 
                       variable=self.auto_split, style='Main.TCheckbutton').grid(row=5, column=0, columnspan=2, sticky="w")
        
        # Right frame for constraints
        right_frame = ttk.LabelFrame(main_frame, text="Constraints", padding="5", style='Main.TLabelframe')
        right_frame.grid(row=0, column=1, sticky="nsew", padx=5, pady=5)
        
        # Precedence constraints
        ttk.Label(right_frame, text="Precedence Constraints:", style='Main.TLabel').grid(row=0, column=0, padx=5, pady=2)
        ttk.Label(right_frame, text="Format: job1 job2 (one per line)\nExample: 1 2 means job1 before job2", 
                 justify=tk.LEFT, style='Main.TLabel').grid(row=1, column=0, padx=5)
        self.precedence = scrolledtext.ScrolledText(right_frame, width=30, height=4)
        self.precedence.grid(row=2, column=0, padx=5, pady=2)
        
        # Resource constraints
        ttk.Label(right_frame, text="Resource Constraints:", style='Main.TLabel').grid(row=3, column=0, padx=5, pady=2)
        ttk.Label(right_frame, text="Format: job machine (one per line)\nExample: 1 2 means job1 on machine2", 
                 justify=tk.LEFT, style='Main.TLabel').grid(row=4, column=0, padx=5)
        self.resources = scrolledtext.ScrolledText(right_frame, width=30, height=4)
        self.resources.grid(row=5, column=0, padx=5, pady=2)
        
        # Temporal constraints
        ttk.Label(right_frame, text="Temporal Constraints:", style='Main.TLabel').grid(row=6, column=0, padx=5, pady=2)
        ttk.Label(right_frame, text="Format: job time (one per line)\nExample: 1 5 means job1 starts after time 5", 
                 justify=tk.LEFT, style='Main.TLabel').grid(row=7, column=0, padx=5)
        self.temporal = scrolledtext.ScrolledText(right_frame, width=30, height=4)
        self.temporal.grid(row=8, column=0, padx=5, pady=2)
        
        # Solve button
        ttk.Button(main_frame, text="Solve", command=self.solve, style='Main.TButton').grid(row=1, column=0, columnspan=2, pady=10)
        
        # Results area
        result_frame = ttk.LabelFrame(main_frame, text="Schedule", padding="5", style='Main.TLabelframe')
        result_frame.grid(row=2, column=0, columnspan=2, sticky="nsew", padx=5, pady=5)
        
        # Matplotlib figure for Gantt chart
        plt.style.use('default')  # Reset to default light style
        self.fig, self.ax = plt.subplots(figsize=(16, 6))
        self.fig.patch.set_facecolor(bg_color)
        self.ax.set_facecolor('white')
        self.canvas = FigureCanvasTkAgg(self.fig, master=result_frame)
        self.canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
        
        # Configure grid weights
        self.root.grid_rowconfigure(0, weight=1)
        self.root.grid_columnconfigure(0, weight=1)
        main_frame.grid_rowconfigure(2, weight=1)
        main_frame.grid_columnconfigure(0, weight=1)
        main_frame.grid_columnconfigure(1, weight=1)
    def plot_schedule(self, schedule, makespan):
        self.ax.clear()
        colors = plt.cm.Set3(np.linspace(0, 1, len(schedule)))
        
        # Set light theme colors
        self.fig.patch.set_facecolor('white')
        self.ax.set_facecolor('white')
        self.ax.spines['bottom'].set_color('black')
        self.ax.spines['top'].set_color('black')
        self.ax.spines['right'].set_color('black')
        self.ax.spines['left'].set_color('black')
        self.ax.tick_params(axis='x', colors='black')
        self.ax.tick_params(axis='y', colors='black')

        # Get machine capacities
        machine_caps = [int(x.strip()) for x in self.machine_caps.get().split(',')]
        num_machines = int(self.num_machines.get())
        
        # Create a dictionary to track jobs at each time slot for each machine
        time_slots = {}
        for job, (start, end, machine) in schedule.items():
            for t in range(start, end):
                if t not in time_slots:
                    time_slots[t] = {}
                if machine not in time_slots[t]:
                    time_slots[t][machine] = []
                time_slots[t][machine].append(job)

        # For each job, find its position within the machine's capacity
        job_positions = {}
        for job, (start, end, machine) in schedule.items():
            # Find concurrent jobs
            concurrent_jobs = set()
            for t in range(start, end):
                if t in time_slots and machine in time_slots[t]:
                    concurrent_jobs.update(time_slots[t][machine])
            
            # Assign position based on concurrent jobs
            used_positions = {job_positions.get(j, -1) for j in concurrent_jobs if j in job_positions}
            pos = 0
            while pos in used_positions:
                pos += 1
            job_positions[job] = pos

        # Plot each job
        for job, (start, end, machine) in schedule.items():
            position = job_positions[job]
            # Calculate y position: base position for machine + slot position
            y_pos = f"Machine {machine+1} (Slot {position+1})"
            
            self.ax.barh(y=y_pos, width=end-start, left=start, 
                        color=colors[job], label=f"Job {job+1}")

        # Adjust y-axis
        self.ax.invert_yaxis()  # Invert to show slots from top to bottom
        
        self.ax.set_xlabel('Time', color='black')
        self.ax.set_title(f'Job Schedule (Makespan: {makespan})', color='black')
        self.ax.grid(True, color='gray', linestyle='-', alpha=0.2)
        
        # Remove duplicate legend entries
        handles, labels = self.ax.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        self.ax.legend(by_label.values(), by_label.keys(), 
                      bbox_to_anchor=(1.02, 1), loc='upper left', 
                      borderaxespad=0)
        
        # Adjust layout to prevent legend cutoff
        self.fig.tight_layout()
        # Add extra space on the right for the legend
        self.fig.subplots_adjust(right=0.85)
        self.canvas.draw()
    def parse_constraints(self, text, convert_to_zero_based=True):
        constraints = []
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    a, b = map(int, line.split())
                    if convert_to_zero_based:
                        constraints.append((a-1, b-1))
                    else:
                        constraints.append((a, b))
                except ValueError:
                    continue
        return constraints
    def parse_temporal_constraints(self, text):
        constraints = {}
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    job, time = map(int, line.split())
                    constraints[job-1] = time
                except ValueError:
                    continue
        return constraints
    def solve(self):
        try:
            # Get input values
            job_durations = [int(x.strip()) for x in self.job_durations.get().split(',')]
            machine_capacities = [int(x.strip()) for x in self.machine_caps.get().split(',')]
            num_machines = int(self.num_machines.get())
            
            # Validate basic inputs
            if len(job_durations) != int(self.num_jobs.get()):
                raise ValueError("Number of job durations doesn't match number of jobs")
            if len(machine_capacities) != num_machines:
                raise ValueError("Number of machine capacities doesn't match number of machines")
            
            # Parse constraints
            precedence_constraints = self.parse_constraints(self.precedence.get("1.0", tk.END))
            resource_constraints = self.parse_constraints(self.resources.get("1.0", tk.END))
            temporal_constraints = self.parse_temporal_constraints(self.temporal.get("1.0", tk.END))
            # Call appropriate algorithm
            if self.algorithm.get() == "Backtracking":
                schedule, makespan = test.backtracking(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            else:
                schedule, makespan = test.genetic_algorithm(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            if schedule:
                self.plot_schedule(schedule, makespan)
            else:
                messagebox.showerror("Error", "No valid schedule found")
        except Exception as e:
            messagebox.showerror("Error", str(e))
if __name__ == "__main__":
    root = tk.Tk()
    welcome = WelcomePage(root)
    root.mainloop() 
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
from PIL import Image, ImageTk
import test

class ModernButton(ttk.Button):
    def __init__(self, master=None, **kwargs):
        super().__init__(master, **kwargs)
        self.bind('<Enter>', self.on_enter)
        self.bind('<Leave>', self.on_leave)
        
    def on_enter(self, e):
        self.configure(style='Modern.TButton.Hover')
        
    def on_leave(self, e):
        self.configure(style='Modern.TButton')

class WelcomePage:
    def __init__(self, root):
        self.root = root
        self.root.title("Job Scheduler - Welcome")
        self.root.state('zoomed')
        
        # Modern color scheme
        self.colors = {
            'bg': '#F8F9FA',  # Light gray background
            'primary': '#4361EE',  # Modern blue
            'secondary': '#3A0CA3',  # Deep purple
            'accent': '#7209B7',  # Purple accent
            'text': '#2B2D42',  # Dark text
            'light_text': '#8D99AE'  # Light text
        }
        
        # Set background color
        self.root.configure(bg=self.colors['bg'])
        
        # Modern style configuration
        style = ttk.Style()
        style.configure('Modern.TFrame', background=self.colors['bg'])
        style.configure('Modern.TLabel', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 12))
        style.configure('Title.TLabel', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 36, 'bold'),
                       foreground=self.colors['primary'])
        style.configure('Subtitle.TLabel', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 18),
                       foreground=self.colors['text'])
        style.configure('Modern.TButton',
                       font=('Segoe UI', 12),
                       padding=15,
                       background=self.colors['primary'],
                       foreground='white')
        style.configure('Modern.TButton.Hover',
                       background=self.colors['secondary'])
        
        # Main frame with modern styling
        main_frame = ttk.Frame(self.root, padding="40", style='Modern.TFrame')
        main_frame.place(relx=0.5, rely=0.5, anchor="center")
        
        # Title with modern styling
        title_frame = ttk.Frame(main_frame, style='Modern.TFrame')
        title_frame.grid(row=0, column=0, pady=(0, 20))
        
        title_label = ttk.Label(title_frame, 
                              text="Job Scheduling Problem Solving",
                              style='Title.TLabel')
        title_label.grid(row=0, column=0, pady=10)
        
        # Modern subtitle
        subtitle_label = ttk.Label(main_frame, 
                                 text="You Want It When?",
                                 style='Subtitle.TLabel',
                                 foreground=self.colors['accent'])
        subtitle_label.grid(row=1, column=0, pady=10)
        
        # Second subtitle with modern styling
        choose_label = ttk.Label(main_frame,
                               text="Choose Your Algorithm",
                               style='Subtitle.TLabel')
        choose_label.grid(row=2, column=0, pady=(30, 20))
        
        # Modern button frame
        button_frame = ttk.Frame(main_frame, style='Modern.TFrame')
        button_frame.grid(row=3, column=0, pady=30)
        
        # Modern algorithm buttons
        backtracking_btn = ModernButton(button_frame, 
                                    text="Backtracking",
                                    style='Modern.TButton',
                                    command=lambda: self.start_scheduler("Backtracking"))
        backtracking_btn.grid(row=0, column=0, padx=20)
        
        genetic_btn = ModernButton(button_frame,
                               text="Genetic Algorithm",
                               style='Modern.TButton',
                               command=lambda: self.start_scheduler("Genetic"))
        genetic_btn.grid(row=0, column=1, padx=20)
        
        # Configure grid weights
        main_frame.grid_columnconfigure(0, weight=1)
        button_frame.grid_columnconfigure((0, 1), weight=1)
        self.root.grid_rowconfigure(0, weight=1)
        self.root.grid_columnconfigure(0, weight=1)
        
    def start_scheduler(self, algorithm):
        self.root.destroy()
        new_root = tk.Tk()
        app = SimpleSchedulerGUI(new_root, initial_algorithm=algorithm)
        new_root.mainloop()

class SimpleSchedulerGUI:
    def __init__(self, root, initial_algorithm="Backtracking"):
        self.root = root
        self.root.title("Simple Job Scheduler")
        self.root.state('zoomed')
        
        # Modern color scheme
        self.colors = {
            'bg': '#F8F9FA',  # Light gray background
            'primary': '#4361EE',  # Modern blue
            'secondary': '#3A0CA3',  # Deep purple
            'accent': '#7209B7',  # Purple accent
            'text': '#2B2D42',  # Dark text
            'light_text': '#8D99AE',  # Light text
            'input_bg': '#FFFFFF',  # White input background
            'border': '#E9ECEF'  # Light border
        }
        
        # Set background color
        self.root.configure(bg=self.colors['bg'])
        
        # Modern style configuration
        style = ttk.Style()
        style.configure('Modern.TFrame', background=self.colors['bg'])
        style.configure('Modern.TLabelframe', 
                       background=self.colors['bg'],
                       borderwidth=2,
                       relief="solid")
        style.configure('Modern.TLabelframe.Label', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 12, 'bold'),
                       foreground=self.colors['primary'])
        style.configure('Modern.TLabel', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 11),
                       foreground=self.colors['text'])
        style.configure('Modern.TButton',
                       font=('Segoe UI', 11),
                       padding=10,
                       background=self.colors['primary'],
                       foreground='white')
        style.configure('Modern.TButton.Hover',
                       background=self.colors['secondary'])
        style.configure('Modern.TEntry',
                       fieldbackground=self.colors['input_bg'],
                       borderwidth=1,
                       relief="solid")
        
        # Main frame with modern styling
        main_frame = ttk.Frame(self.root, padding="20", style='Modern.TFrame')
        main_frame.grid(row=0, column=0, sticky="nsew")
        
        # Left frame for basic inputs with modern styling
        left_frame = ttk.LabelFrame(main_frame, 
                                  text="Basic Data", 
                                  padding="15", 
                                  style='Modern.TLabelframe')
        left_frame.grid(row=0, column=0, sticky="nsew", padx=10, pady=10)
        
        # Modern input fields
        ttk.Label(left_frame, 
                 text="Number of Jobs:", 
                 style='Modern.TLabel').grid(row=0, column=0, padx=5, pady=5)
        self.num_jobs = ttk.Entry(left_frame, 
                                width=10, 
                                style='Modern.TEntry')
        self.num_jobs.grid(row=0, column=1, padx=5, pady=5)
        self.num_jobs.insert(0, "2")
        
        ttk.Label(left_frame, 
                 text="Job Durations (comma-separated):", 
                 style='Modern.TLabel').grid(row=1, column=0, padx=5, pady=5)
        self.job_durations = ttk.Entry(left_frame, 
                                     width=30, 
                                     style='Modern.TEntry')
        self.job_durations.grid(row=1, column=1, padx=5, pady=5)
        self.job_durations.insert(0, "2,3")
        
        # Machine inputs
        ttk.Label(left_frame, text="Number of Machines:", style='Modern.TLabel').grid(row=2, column=0, padx=5, pady=5)
        self.num_machines = ttk.Entry(left_frame, width=10, style='Modern.TEntry')
        self.num_machines.grid(row=2, column=1, padx=5, pady=5)
        self.num_machines.insert(0, "2")
        
        ttk.Label(left_frame, text="Machine Capacities (comma-separated):", style='Modern.TLabel').grid(row=3, column=0, padx=5, pady=5)
        self.machine_caps = ttk.Entry(left_frame, width=30, style='Modern.TEntry')
        self.machine_caps.grid(row=3, column=1, padx=5, pady=5)
        self.machine_caps.insert(0, "1,1")
        
        # Algorithm selection
        ttk.Label(left_frame, text="Algorithm:", style='Modern.TLabel').grid(row=4, column=0, padx=5, pady=5)
        self.algorithm = ttk.Combobox(left_frame, values=["Backtracking", "Genetic"], state="readonly", style='Modern.TEntry')
        self.algorithm.grid(row=4, column=1, padx=5, pady=5)
        self.algorithm.set(initial_algorithm)
        
        # Auto-split option
        self.auto_split = tk.BooleanVar(value=True)
        ttk.Checkbutton(left_frame, text="Auto-assign jobs to machines", 
                       variable=self.auto_split, style='Modern.TCheckbutton').grid(row=5, column=0, columnspan=2, sticky="w")
        
        # Right frame for constraints
        right_frame = ttk.LabelFrame(main_frame, text="Constraints", padding="15", style='Modern.TLabelframe')
        right_frame.grid(row=1, column=0, sticky="nsew", padx=10, pady=10)
        
        # Precedence constraints
        ttk.Label(right_frame, text="Precedence Constraints:", style='Modern.TLabel').grid(row=0, column=0, padx=5, pady=5)
        ttk.Label(right_frame, text="Format: job1 job2 (one per line)\nExample: 1 2 means job1 before job2", 
                 justify=tk.LEFT, style='Modern.TLabel').grid(row=1, column=0, padx=5)
        self.precedence = scrolledtext.ScrolledText(right_frame, width=30, height=4, style='Modern.TEntry')
        self.precedence.grid(row=2, column=0, padx=5, pady=5)
        
        # Resource constraints
        ttk.Label(right_frame, text="Resource Constraints:", style='Modern.TLabel').grid(row=3, column=0, padx=5, pady=5)
        ttk.Label(right_frame, text="Format: job machine (one per line)\nExample: 1 2 means job1 on machine2", 
                 justify=tk.LEFT, style='Modern.TLabel').grid(row=4, column=0, padx=5)
        self.resources = scrolledtext.ScrolledText(right_frame, width=30, height=4, style='Modern.TEntry')
        self.resources.grid(row=5, column=0, padx=5, pady=5)
        
        # Temporal constraints
        ttk.Label(right_frame, text="Temporal Constraints:", style='Modern.TLabel').grid(row=6, column=0, padx=5, pady=5)
        ttk.Label(right_frame, text="Format: job time (one per line)\nExample: 1 5 means job1 starts after time 5", 
                 justify=tk.LEFT, style='Modern.TLabel').grid(row=7, column=0, padx=5)
        self.temporal = scrolledtext.ScrolledText(right_frame, width=30, height=4, style='Modern.TEntry')
        self.temporal.grid(row=8, column=0, padx=5, pady=5)
        
        # Modern solve button
        solve_btn = ModernButton(main_frame, 
                               text="Solve", 
                               command=self.solve, 
                               style='Modern.TButton')
        solve_btn.grid(row=2, column=0, columnspan=2, pady=20)
        
        # Results area with modern styling
        result_frame = ttk.LabelFrame(main_frame, 
                                    text="Schedule", 
                                    padding="15", 
                                    style='Modern.TLabelframe')
        result_frame.grid(row=3, column=0, columnspan=2, sticky="nsew", padx=10, pady=10)
        
        # Modern matplotlib figure
        plt.style.use('seaborn')  # Modern matplotlib style
        self.fig, self.ax = plt.subplots(figsize=(16, 6))
        self.fig.patch.set_facecolor(self.colors['bg'])
        self.ax.set_facecolor('white')
        self.canvas = FigureCanvasTkAgg(self.fig, master=result_frame)
        self.canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
        
        # Configure grid weights
        self.root.grid_rowconfigure(0, weight=1)
        self.root.grid_columnconfigure(0, weight=1)
        main_frame.grid_rowconfigure(3, weight=1)
        main_frame.grid_columnconfigure(0, weight=1)
        main_frame.grid_columnconfigure(1, weight=1)
    def plot_schedule(self, schedule, makespan):
        self.ax.clear()
        colors = plt.cm.Set3(np.linspace(0, 1, len(schedule)))
        
        # Set light theme colors
        self.fig.patch.set_facecolor('white')
        self.ax.set_facecolor('white')
        self.ax.spines['bottom'].set_color('black')
        self.ax.spines['top'].set_color('black')
        self.ax.spines['right'].set_color('black')
        self.ax.spines['left'].set_color('black')
        self.ax.tick_params(axis='x', colors='black')
        self.ax.tick_params(axis='y', colors='black')

        # Get machine capacities
        machine_caps = [int(x.strip()) for x in self.machine_caps.get().split(',')]
        num_machines = int(self.num_machines.get())
        
        # Create a dictionary to track jobs at each time slot for each machine
        time_slots = {}
        for job, (start, end, machine) in schedule.items():
            for t in range(start, end):
                if t not in time_slots:
                    time_slots[t] = {}
                if machine not in time_slots[t]:
                    time_slots[t][machine] = []
                time_slots[t][machine].append(job)

        # For each job, find its position within the machine's capacity
        job_positions = {}
        for job, (start, end, machine) in schedule.items():
            # Find concurrent jobs
            concurrent_jobs = set()
            for t in range(start, end):
                if t in time_slots and machine in time_slots[t]:
                    concurrent_jobs.update(time_slots[t][machine])
            
            # Assign position based on concurrent jobs
            used_positions = {job_positions.get(j, -1) for j in concurrent_jobs if j in job_positions}
            pos = 0
            while pos in used_positions:
                pos += 1
            job_positions[job] = pos

        # Plot each job
        for job, (start, end, machine) in schedule.items():
            position = job_positions[job]
            # Calculate y position: base position for machine + slot position
            y_pos = f"Machine {machine+1} (Slot {position+1})"
            
            self.ax.barh(y=y_pos, width=end-start, left=start, 
                        color=colors[job], label=f"Job {job+1}")

        # Adjust y-axis
        self.ax.invert_yaxis()  # Invert to show slots from top to bottom
        
        self.ax.set_xlabel('Time', color='black')
        self.ax.set_title(f'Job Schedule (Makespan: {makespan})', color='black')
        self.ax.grid(True, color='gray', linestyle='-', alpha=0.2)
        
        # Remove duplicate legend entries
        handles, labels = self.ax.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        self.ax.legend(by_label.values(), by_label.keys(), 
                      bbox_to_anchor=(1.02, 1), loc='upper left', 
                      borderaxespad=0)
        
        # Adjust layout to prevent legend cutoff
        self.fig.tight_layout()
        # Add extra space on the right for the legend
        self.fig.subplots_adjust(right=0.85)
        self.canvas.draw()
    def parse_constraints(self, text, convert_to_zero_based=True):
        constraints = []
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    a, b = map(int, line.split())
                    if convert_to_zero_based:
                        constraints.append((a-1, b-1))
                    else:
                        constraints.append((a, b))
                except ValueError:
                    continue
        return constraints
    def parse_temporal_constraints(self, text):
        constraints = {}
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    job, time = map(int, line.split())
                    constraints[job-1] = time
                except ValueError:
                    continue
        return constraints
    def solve(self):
        try:
            # Get input values
            job_durations = [int(x.strip()) for x in self.job_durations.get().split(',')]
            machine_capacities = [int(x.strip()) for x in self.machine_caps.get().split(',')]
            num_machines = int(self.num_machines.get())
            
            # Validate basic inputs
            if len(job_durations) != int(self.num_jobs.get()):
                raise ValueError("Number of job durations doesn't match number of jobs")
            if len(machine_capacities) != num_machines:
                raise ValueError("Number of machine capacities doesn't match number of machines")
            
            # Parse constraints
            precedence_constraints = self.parse_constraints(self.precedence.get("1.0", tk.END))
            resource_constraints = self.parse_constraints(self.resources.get("1.0", tk.END))
            temporal_constraints = self.parse_temporal_constraints(self.temporal.get("1.0", tk.END))
            # Call appropriate algorithm
            if self.algorithm.get() == "Backtracking":
                schedule, makespan = test.backtracking(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            else:
                schedule, makespan = test.genetic_algorithm(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            if schedule:
                self.plot_schedule(schedule, makespan)
            else:
                messagebox.showerror("Error", "No valid schedule found")
        except Exception as e:
            messagebox.showerror("Error", str(e))
if __name__ == "__main__":
    root = tk.Tk()
    welcome = WelcomePage(root)
    root.mainloop() 
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
from PIL import Image, ImageTk
import test

class ModernButton(ttk.Button):
    def __init__(self, master=None, **kwargs):
        super().__init__(master, **kwargs)
        self.bind('<Enter>', self.on_enter)
        self.bind('<Leave>', self.on_leave)
        
    def on_enter(self, e):
        self.configure(style='Modern.TButton.Hover')
        
    def on_leave(self, e):
        self.configure(style='Modern.TButton')

class WelcomePage:
    def __init__(self, root):
        self.root = root
        self.root.title("Job Scheduler - Welcome")
        self.root.state('zoomed')
        
        # Modern color scheme
        self.colors = {
            'bg': '#F8F9FA',  # Light gray background
            'primary': '#4361EE',  # Modern blue
            'secondary': '#3A0CA3',  # Deep purple
            'accent': '#7209B7',  # Purple accent
            'text': '#2B2D42',  # Dark text
            'light_text': '#8D99AE'  # Light text
        }
        
        # Set background color
        self.root.configure(bg=self.colors['bg'])
        
        # Modern style configuration
        style = ttk.Style()
        style.configure('Modern.TFrame', background=self.colors['bg'])
        style.configure('Modern.TLabel', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 12))
        style.configure('Title.TLabel', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 36, 'bold'),
                       foreground=self.colors['primary'])
        style.configure('Subtitle.TLabel', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 18),
                       foreground=self.colors['text'])
        style.configure('Modern.TButton',
                       font=('Segoe UI', 12),
                       padding=15,
                       background=self.colors['primary'],
                       foreground='white')
        style.configure('Modern.TButton.Hover',
                       background=self.colors['secondary'])
        
        # Main frame with modern styling
        main_frame = ttk.Frame(self.root, padding="40", style='Modern.TFrame')
        main_frame.place(relx=0.5, rely=0.5, anchor="center")
        
        # Title with modern styling
        title_frame = ttk.Frame(main_frame, style='Modern.TFrame')
        title_frame.grid(row=0, column=0, pady=(0, 20))
        
        title_label = ttk.Label(title_frame, 
                              text="Job Scheduling Problem Solving",
                              style='Title.TLabel')
        title_label.grid(row=0, column=0, pady=10)
        
        # Modern subtitle
        subtitle_label = ttk.Label(main_frame, 
                                 text="You Want It When?",
                                 style='Subtitle.TLabel',
                                 foreground=self.colors['accent'])
        subtitle_label.grid(row=1, column=0, pady=10)
        
        # Second subtitle with modern styling
        choose_label = ttk.Label(main_frame,
                               text="Choose Your Algorithm",
                               style='Subtitle.TLabel')
        choose_label.grid(row=2, column=0, pady=(30, 20))
        
        # Modern button frame
        button_frame = ttk.Frame(main_frame, style='Modern.TFrame')
        button_frame.grid(row=3, column=0, pady=30)
        
        # Modern algorithm buttons
        backtracking_btn = ModernButton(button_frame, 
                                    text="Backtracking",
                                    style='Modern.TButton',
                                    command=lambda: self.start_scheduler("Backtracking"))
        backtracking_btn.grid(row=0, column=0, padx=20)
        
        genetic_btn = ModernButton(button_frame,
                               text="Genetic Algorithm",
                               style='Modern.TButton',
                               command=lambda: self.start_scheduler("Genetic"))
        genetic_btn.grid(row=0, column=1, padx=20)
        
        # Configure grid weights
        main_frame.grid_columnconfigure(0, weight=1)
        button_frame.grid_columnconfigure((0, 1), weight=1)
        self.root.grid_rowconfigure(0, weight=1)
        self.root.grid_columnconfigure(0, weight=1)
        
    def start_scheduler(self, algorithm):
        self.root.destroy()
        new_root = tk.Tk()
        app = SimpleSchedulerGUI(new_root, initial_algorithm=algorithm)
        new_root.mainloop()

class SimpleSchedulerGUI:
    def __init__(self, root, initial_algorithm="Backtracking"):
        self.root = root
        self.root.title("Simple Job Scheduler")
        self.root.state('zoomed')
        
        # Modern color scheme
        self.colors = {
            'bg': '#F8F9FA',  # Light gray background
            'primary': '#4361EE',  # Modern blue
            'secondary': '#3A0CA3',  # Deep purple
            'accent': '#7209B7',  # Purple accent
            'text': '#2B2D42',  # Dark text
            'light_text': '#8D99AE',  # Light text
            'input_bg': '#FFFFFF',  # White input background
            'border': '#E9ECEF'  # Light border
        }
        
        # Set background color
        self.root.configure(bg=self.colors['bg'])
        
        # TEST LABEL to confirm page loads
        test_label = tk.Label(self.root, text="Scheduler Page Loaded", font=("Segoe UI", 16), bg=self.colors['bg'], fg=self.colors['primary'])
        test_label.grid(row=0, column=0, pady=10)
        
        # Modern style configuration
        style = ttk.Style()
        style.configure('Modern.TFrame', background=self.colors['bg'])
        style.configure('Modern.TLabelframe', 
                       background=self.colors['bg'],
                       borderwidth=2,
                       relief="solid")
        style.configure('Modern.TLabelframe.Label', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 12, 'bold'),
                       foreground=self.colors['primary'])
        style.configure('Modern.TLabel', 
                       background=self.colors['bg'],
                       font=('Segoe UI', 11),
                       foreground=self.colors['text'])
        style.configure('Modern.TButton',
                       font=('Segoe UI', 11),
                       padding=10,
                       background=self.colors['primary'],
                       foreground='white')
        style.configure('Modern.TButton.Hover',
                       background=self.colors['secondary'])
        style.configure('Modern.TEntry',
                       fieldbackground=self.colors['input_bg'],
                       borderwidth=1,
                       relief="solid")
        
        # Add modern styles for combobox and checkbutton
        style.configure('Modern.TCombobox',
                       fieldbackground=self.colors['input_bg'],
                       background=self.colors['input_bg'],
                       foreground=self.colors['text'],
                       arrowcolor=self.colors['primary'],
                       borderwidth=1,
                       relief="solid")
        
        style.configure('Modern.TCheckbutton',
                       background=self.colors['bg'],
                       foreground=self.colors['text'],
                       font=('Segoe UI', 11))
        
        # Configure scrolledtext styling
        style.configure('Modern.TScrolledText',
                       fieldbackground=self.colors['input_bg'],
                       foreground=self.colors['text'],
                       borderwidth=1,
                       relief="solid")
        
        # Add focus effects for entry widgets
        def on_focus_in(event):
            event.widget.configure(style='Modern.TEntry.Focus')
            
        def on_focus_out(event):
            event.widget.configure(style='Modern.TEntry')
        
        # Configure focused entry style
        style.configure('Modern.TEntry.Focus',
                       fieldbackground=self.colors['input_bg'],
                       borderwidth=2,
                       relief="solid")
        
        # Apply focus effects to all entry widgets
        for widget in [self.num_jobs, self.job_durations, self.num_machines, self.machine_caps]:
            widget.bind('<FocusIn>', on_focus_in)
            widget.bind('<FocusOut>', on_focus_out)
        
        # Style the combobox
        self.algorithm.configure(style='Modern.TCombobox')
        
        # Style the scrolledtext widgets
        for widget in [self.precedence, self.resources, self.temporal]:
            widget.configure(
                font=('Segoe UI', 11),
                bg=self.colors['input_bg'],
                fg=self.colors['text'],
                insertbackground=self.colors['primary'],
                selectbackground=self.colors['primary'],
                selectforeground='white',
                relief="solid",
                borderwidth=1
            )
        
        # Main frame with modern styling
        main_frame = ttk.Frame(self.root, padding="20", style='Modern.TFrame')
        main_frame.grid(row=1, column=0, sticky="nsew")
        
        # Left frame for basic inputs with modern styling
        left_frame = ttk.LabelFrame(main_frame, 
                                  text="Basic Data", 
                                  padding="15", 
                                  style='Modern.TLabelframe')
        left_frame.grid(row=0, column=0, sticky="nsew", padx=10, pady=10)
        
        # Modern input fields
        ttk.Label(left_frame, 
                 text="Number of Jobs:", 
                 style='Modern.TLabel').grid(row=0, column=0, padx=5, pady=5)
        self.num_jobs = ttk.Entry(left_frame, 
                                width=10, 
                                style='Modern.TEntry')
        self.num_jobs.grid(row=0, column=1, padx=5, pady=5)
        self.num_jobs.insert(0, "2")
        
        ttk.Label(left_frame, 
                 text="Job Durations (comma-separated):", 
                 style='Modern.TLabel').grid(row=1, column=0, padx=5, pady=5)
        self.job_durations = ttk.Entry(left_frame, 
                                     width=30, 
                                     style='Modern.TEntry')
        self.job_durations.grid(row=1, column=1, padx=5, pady=5)
        self.job_durations.insert(0, "2,3")
        
        # Machine inputs
        ttk.Label(left_frame, text="Number of Machines:", style='Modern.TLabel').grid(row=2, column=0, padx=5, pady=5)
        self.num_machines = ttk.Entry(left_frame, width=10, style='Modern.TEntry')
        self.num_machines.grid(row=2, column=1, padx=5, pady=5)
        self.num_machines.insert(0, "2")
        
        ttk.Label(left_frame, text="Machine Capacities (comma-separated):", style='Modern.TLabel').grid(row=3, column=0, padx=5, pady=5)
        self.machine_caps = ttk.Entry(left_frame, width=30, style='Modern.TEntry')
        self.machine_caps.grid(row=3, column=1, padx=5, pady=5)
        self.machine_caps.insert(0, "1,1")
        
        # Algorithm selection
        ttk.Label(left_frame, text="Algorithm:", style='Modern.TLabel').grid(row=4, column=0, padx=5, pady=5)
        self.algorithm = ttk.Combobox(left_frame, values=["Backtracking", "Genetic"], state="readonly", style='Modern.TEntry')
        self.algorithm.grid(row=4, column=1, padx=5, pady=5)
        self.algorithm.set(initial_algorithm)
        
        # Auto-split option
        self.auto_split = tk.BooleanVar(value=True)
        ttk.Checkbutton(left_frame, text="Auto-assign jobs to machines", 
                       variable=self.auto_split, style='Modern.TCheckbutton').grid(row=5, column=0, columnspan=2, sticky="w")
        
        # Right frame for constraints
        right_frame = ttk.LabelFrame(main_frame, text="Constraints", padding="15", style='Modern.TLabelframe')
        right_frame.grid(row=1, column=0, sticky="nsew", padx=10, pady=10)
        
        # Precedence constraints
        ttk.Label(right_frame, text="Precedence Constraints:", style='Modern.TLabel').grid(row=0, column=0, padx=5, pady=5)
        ttk.Label(right_frame, text="Format: job1 job2 (one per line)\nExample: 1 2 means job1 before job2", 
                 justify=tk.LEFT, style='Modern.TLabel').grid(row=1, column=0, padx=5)
        self.precedence = scrolledtext.ScrolledText(right_frame, width=30, height=4, style='Modern.TEntry')
        self.precedence.grid(row=2, column=0, padx=5, pady=5)
        
        # Resource constraints
        ttk.Label(right_frame, text="Resource Constraints:", style='Modern.TLabel').grid(row=3, column=0, padx=5, pady=5)
        ttk.Label(right_frame, text="Format: job machine (one per line)\nExample: 1 2 means job1 on machine2", 
                 justify=tk.LEFT, style='Modern.TLabel').grid(row=4, column=0, padx=5)
        self.resources = scrolledtext.ScrolledText(right_frame, width=30, height=4, style='Modern.TEntry')
        self.resources.grid(row=5, column=0, padx=5, pady=5)
        
        # Temporal constraints
        ttk.Label(right_frame, text="Temporal Constraints:", style='Modern.TLabel').grid(row=6, column=0, padx=5, pady=5)
        ttk.Label(right_frame, text="Format: job time (one per line)\nExample: 1 5 means job1 starts after time 5", 
                 justify=tk.LEFT, style='Modern.TLabel').grid(row=7, column=0, padx=5)
        self.temporal = scrolledtext.ScrolledText(right_frame, width=30, height=4, style='Modern.TEntry')
        self.temporal.grid(row=8, column=0, padx=5, pady=5)
        
        # Modern solve button
        solve_btn = ModernButton(main_frame, 
                               text="Solve", 
                               command=self.solve, 
                               style='Modern.TButton')
        solve_btn.grid(row=2, column=0, columnspan=2, pady=20)
        
        # Results area with modern styling
        result_frame = ttk.LabelFrame(main_frame, 
                                    text="Schedule", 
                                    padding="15", 
                                    style='Modern.TLabelframe')
        result_frame.grid(row=3, column=0, columnspan=2, sticky="nsew", padx=10, pady=10)
        
        # Modern matplotlib figure
        plt.style.use('seaborn')  # Modern matplotlib style
        self.fig, self.ax = plt.subplots(figsize=(16, 6))
        self.fig.patch.set_facecolor(self.colors['bg'])
        self.ax.set_facecolor('white')
        self.canvas = FigureCanvasTkAgg(self.fig, master=result_frame)
        self.canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
        
        # Configure grid weights for root and main_frame
        self.root.grid_rowconfigure(0, weight=0)  # test label
        self.root.grid_rowconfigure(1, weight=1)  # main_frame
        self.root.grid_columnconfigure(0, weight=1)
        main_frame.grid_rowconfigure(0, weight=1)
        main_frame.grid_columnconfigure(0, weight=1)
        main_frame.grid_columnconfigure(1, weight=1)
    def plot_schedule(self, schedule, makespan):
        self.ax.clear()
        colors = plt.cm.Set3(np.linspace(0, 1, len(schedule)))
        
        # Set modern theme colors
        self.fig.patch.set_facecolor(self.colors['bg'])
        self.ax.set_facecolor('white')
        self.ax.spines['bottom'].set_color(self.colors['border'])
        self.ax.spines['top'].set_color(self.colors['border'])
        self.ax.spines['right'].set_color(self.colors['border'])
        self.ax.spines['left'].set_color(self.colors['border'])
        self.ax.tick_params(axis='x', colors=self.colors['text'])
        self.ax.tick_params(axis='y', colors=self.colors['text'])

        # Get machine capacities
        machine_caps = [int(x.strip()) for x in self.machine_caps.get().split(',')]
        num_machines = int(self.num_machines.get())
        
        # Create a dictionary to track jobs at each time slot for each machine
        time_slots = {}
        for job, (start, end, machine) in schedule.items():
            for t in range(start, end):
                if t not in time_slots:
                    time_slots[t] = {}
                if machine not in time_slots[t]:
                    time_slots[t][machine] = []
                time_slots[t][machine].append(job)

        # For each job, find its position within the machine's capacity
        job_positions = {}
        for job, (start, end, machine) in schedule.items():
            # Find concurrent jobs
            concurrent_jobs = set()
            for t in range(start, end):
                if t in time_slots and machine in time_slots[t]:
                    concurrent_jobs.update(time_slots[t][machine])
            
            # Assign position based on concurrent jobs
            used_positions = {job_positions.get(j, -1) for j in concurrent_jobs if j in job_positions}
            pos = 0
            while pos in used_positions:
                pos += 1
            job_positions[job] = pos

        # Plot each job with modern styling
        for job, (start, end, machine) in schedule.items():
            position = job_positions[job]
            y_pos = f"Machine {machine+1} (Slot {position+1})"
            
            # Create modern bar style
            bar = self.ax.barh(y=y_pos, width=end-start, left=start, 
                             color=colors[job], label=f"Job {job+1}",
                             alpha=0.8, edgecolor=self.colors['border'],
                             linewidth=1)

        # Modern axis styling
        self.ax.invert_yaxis()
        self.ax.set_xlabel('Time', color=self.colors['text'], fontsize=12)
        self.ax.set_title(f'Job Schedule (Makespan: {makespan})', 
                         color=self.colors['primary'], 
                         fontsize=14, 
                         pad=20)
        
        # Modern grid styling
        self.ax.grid(True, color=self.colors['border'], linestyle='--', alpha=0.3)
        
        # Modern legend styling
        handles, labels = self.ax.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        legend = self.ax.legend(by_label.values(), by_label.keys(), 
                              bbox_to_anchor=(1.02, 1), 
                              loc='upper left',
                              borderaxespad=0,
                              frameon=True,
                              facecolor='white',
                              edgecolor=self.colors['border'])
        
        # Adjust layout
        self.fig.tight_layout()
        self.fig.subplots_adjust(right=0.85)
        self.canvas.draw()
    def parse_constraints(self, text, convert_to_zero_based=True):
        constraints = []
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    a, b = map(int, line.split())
                    if convert_to_zero_based:
                        constraints.append((a-1, b-1))
                    else:
                        constraints.append((a, b))
                except ValueError:
                    continue
        return constraints
    def parse_temporal_constraints(self, text):
        constraints = {}
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    job, time = map(int, line.split())
                    constraints[job-1] = time
                except ValueError:
                    continue
        return constraints
    def solve(self):
        try:
            # Get input values
            job_durations = [int(x.strip()) for x in self.job_durations.get().split(',')]
            machine_capacities = [int(x.strip()) for x in self.machine_caps.get().split(',')]
            num_machines = int(self.num_machines.get())
            
            # Validate basic inputs
            if len(job_durations) != int(self.num_jobs.get()):
                raise ValueError("Number of job durations doesn't match number of jobs")
            if len(machine_capacities) != num_machines:
                raise ValueError("Number of machine capacities doesn't match number of machines")
            
            # Parse constraints
            precedence_constraints = self.parse_constraints(self.precedence.get("1.0", tk.END))
            resource_constraints = self.parse_constraints(self.resources.get("1.0", tk.END))
            temporal_constraints = self.parse_temporal_constraints(self.temporal.get("1.0", tk.END))
            # Call appropriate algorithm
            if self.algorithm.get() == "Backtracking":
                schedule, makespan = test.backtracking(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            else:
                schedule, makespan = test.genetic_algorithm(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            if schedule:
                self.plot_schedule(schedule, makespan)
            else:
                messagebox.showerror("Error", "No valid schedule found")
        except Exception as e:
            messagebox.showerror("Error", str(e))
if __name__ == "__main__":
    root = tk.Tk()
    welcome = WelcomePage(root)
    root.mainloop() 
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
import test

# Modern color palette
COLORS = {
    'bg': '#F4F6FB',
    'primary': '#2563eb',
    'accent': '#22d3ee',
    'text': '#22223b',
    'card': '#fff',
    'button': '#2563eb',
    'button_hover': '#1e40af',
    'error': '#ef4444',
    'success': '#22c55e',
}

FONT = ('Segoe UI', 11)
TITLE_FONT = ('Segoe UI', 28, 'bold')
SUBTITLE_FONT = ('Segoe UI', 16)

class ScrollableFrame(ttk.Frame):
    def __init__(self, container, *args, **kwargs):
        super().__init__(container, *args, **kwargs)
        canvas = tk.Canvas(self, bg=COLORS['bg'], highlightthickness=0)
        scrollbar = ttk.Scrollbar(self, orient="vertical", command=canvas.yview)
        self.scrollable_frame = ttk.Frame(canvas, style='Card.TFrame')
        self.scrollable_frame.bind(
            "<Configure>",
            lambda e: canvas.configure(
                scrollregion=canvas.bbox("all")
            )
        )
        canvas.create_window((0, 0), window=self.scrollable_frame, anchor="nw")
        canvas.configure(yscrollcommand=scrollbar.set)
        canvas.pack(side="left", fill="both", expand=True)
        scrollbar.pack(side="right", fill="y")
        self.canvas = canvas
        self.bind_mousewheel(canvas)
    def bind_mousewheel(self, widget):
        widget.bind_all("<MouseWheel>", self._on_mousewheel)
    def _on_mousewheel(self, event):
        self.canvas.yview_scroll(int(-1*(event.delta/120)), "units")

class ModernButton(ttk.Button):
    def __init__(self, master=None, **kwargs):
        super().__init__(master, **kwargs)
        self.default_style = kwargs.get('style', 'Modern.TButton')
        self.hover_style = 'Modern.TButton.Hover'
        self.bind('<Enter>', self.on_enter)
        self.bind('<Leave>', self.on_leave)
    def on_enter(self, e):
        self.configure(style=self.hover_style)
    def on_leave(self, e):
        self.configure(style=self.default_style)

class WelcomePage:
    def __init__(self, root, switch_callback):
        self.root = root
        self.switch_callback = switch_callback
        self.root.configure(bg=COLORS['bg'])
        style = ttk.Style()
        style.configure('Card.TFrame', background=COLORS['card'], relief='flat')
        style.configure('Title.TLabel', background=COLORS['card'], font=TITLE_FONT, foreground=COLORS['primary'])
        style.configure('Subtitle.TLabel', background=COLORS['card'], font=SUBTITLE_FONT, foreground=COLORS['text'])
        style.configure('Modern.TButton', font=FONT, background=COLORS['button'], foreground='white', padding=16)
        style.configure('Modern.TButton.Hover', background=COLORS['button_hover'], foreground='white')
        # Centered card
        card = ttk.Frame(self.root, style='Card.TFrame', padding=40)
        card.place(relx=0.5, rely=0.5, anchor='center')
        ttk.Label(card, text='Job Scheduler', style='Title.TLabel').pack(pady=(0, 10))
        ttk.Label(card, text='Optimize your workflow', style='Subtitle.TLabel').pack(pady=(0, 30))
        ModernButton(card, text='Backtracking', command=lambda: self.switch_callback('Backtracking')).pack(fill='x', pady=8)
        ModernButton(card, text='Genetic Algorithm', command=lambda: self.switch_callback('Genetic')).pack(fill='x', pady=8)
        ttk.Label(card, text='by Your Name', style='Subtitle.TLabel', font=('Segoe UI', 10), foreground=COLORS['accent']).pack(pady=(30,0))

class SchedulerPage:
    def __init__(self, root, algorithm, back_callback):
        self.root = root
        self.algorithm = algorithm
        self.back_callback = back_callback
        self.root.configure(bg=COLORS['bg'])
        style = ttk.Style()
        style.configure('Card.TFrame', background=COLORS['card'], relief='flat')
        style.configure('Section.TLabelframe', background=COLORS['card'], font=FONT)
        style.configure('Section.TLabelframe.Label', background=COLORS['card'], font=FONT, foreground=COLORS['primary'])
        style.configure('Modern.TButton', font=FONT, background=COLORS['button'], foreground='white', padding=12)
        style.configure('Modern.TButton.Hover', background=COLORS['button_hover'], foreground='white')
        style.configure('Modern.TEntry', fieldbackground=COLORS['card'], borderwidth=1, relief='solid')
        # Header
        header = ttk.Frame(self.root, style='Card.TFrame', padding=(30, 20))
        header.pack(fill='x', pady=(0, 10))
        ttk.Label(header, text='Configure Your Job Scheduling Problem', style='Title.TLabel').pack(side='left')
        ModernButton(header, text='Back', command=self.back_callback, style='Modern.TButton').pack(side='right')
        # Main content (now scrollable)
        scrollable = ScrollableFrame(self.root)
        scrollable.pack(fill='both', expand=True)
        content = scrollable.scrollable_frame
        # Input section
        input_card = ttk.Labelframe(content, text='Basic Data', style='Section.TLabelframe', padding=20)
        input_card.grid(row=0, column=0, sticky='nsew', padx=10, pady=10)
        ttk.Label(input_card, text='Number of Jobs:', style='Subtitle.TLabel').grid(row=0, column=0, sticky='w', pady=4)
        self.num_jobs = ttk.Entry(input_card, width=10, style='Modern.TEntry')
        self.num_jobs.grid(row=0, column=1, pady=4)
        self.num_jobs.insert(0, '2')
        ttk.Label(input_card, text='Job Durations (comma-separated):', style='Subtitle.TLabel').grid(row=1, column=0, sticky='w', pady=4)
        self.job_durations = ttk.Entry(input_card, width=30, style='Modern.TEntry')
        self.job_durations.grid(row=1, column=1, pady=4)
        self.job_durations.insert(0, '2,3')
        ttk.Label(input_card, text='Number of Machines:', style='Subtitle.TLabel').grid(row=2, column=0, sticky='w', pady=4)
        self.num_machines = ttk.Entry(input_card, width=10, style='Modern.TEntry')
        self.num_machines.grid(row=2, column=1, pady=4)
        self.num_machines.insert(0, '2')
        ttk.Label(input_card, text='Machine Capacities (comma-separated):', style='Subtitle.TLabel').grid(row=3, column=0, sticky='w', pady=4)
        self.machine_caps = ttk.Entry(input_card, width=30, style='Modern.TEntry')
        self.machine_caps.grid(row=3, column=1, pady=4)
        self.machine_caps.insert(0, '1,1')
        # Algorithm selection (show as label, but keep for logic)
        ttk.Label(input_card, text='Algorithm:', style='Subtitle.TLabel').grid(row=4, column=0, sticky='w', pady=4)
        self.algorithm_label = ttk.Label(input_card, text=algorithm, style='Subtitle.TLabel', foreground=COLORS['primary'])
        self.algorithm_label.grid(row=4, column=1, pady=4, sticky='w')
        # Auto-assign option
        self.auto_split = tk.BooleanVar(value=True)
        ttk.Checkbutton(input_card, text="Auto-assign jobs to machines", variable=self.auto_split, style='Section.TLabelframe').grid(row=5, column=0, columnspan=2, sticky="w", pady=4)
        # Constraints section
        constraints_card = ttk.Labelframe(content, text='Constraints', style='Section.TLabelframe', padding=20)
        constraints_card.grid(row=1, column=0, sticky='nsew', padx=10, pady=10)
        ttk.Label(constraints_card, text='Precedence Constraints (job1 job2 per line):', style='Subtitle.TLabel').pack(anchor='w', pady=(0,2))
        self.precedence = scrolledtext.ScrolledText(constraints_card, width=30, height=2, font=FONT)
        self.precedence.pack(fill='x', pady=2)
        ttk.Label(constraints_card, text='Resource Constraints (job machine per line):', style='Subtitle.TLabel').pack(anchor='w', pady=(8,2))
        self.resources = scrolledtext.ScrolledText(constraints_card, width=30, height=2, font=FONT)
        self.resources.pack(fill='x', pady=2)
        ttk.Label(constraints_card, text='Temporal Constraints (job time per line):', style='Subtitle.TLabel').pack(anchor='w', pady=(8,2))
        self.temporal = scrolledtext.ScrolledText(constraints_card, width=30, height=2, font=FONT)
        self.temporal.pack(fill='x', pady=2)
        # Solve button
        ModernButton(content, text='Solve', command=self.solve, style='Modern.TButton').grid(row=2, column=0, pady=20, sticky='ew')
        # Results area
        self.result_card = ttk.Labelframe(content, text='Schedule', style='Section.TLabelframe', padding=20)
        self.result_card.grid(row=3, column=0, sticky='nsew', padx=10, pady=10)
        self.result_label = ttk.Label(self.result_card, text='', style='Subtitle.TLabel')
        self.result_label.pack()
        self.fig, self.ax = plt.subplots(figsize=(10, 3))
        self.canvas = FigureCanvasTkAgg(self.fig, master=self.result_card)
        self.canvas.get_tk_widget().pack(fill='both', expand=True)
        # Configure grid weights
        content.grid_rowconfigure(3, weight=1)
        content.grid_columnconfigure(0, weight=1)
    def parse_constraints(self, text, convert_to_zero_based=True):
        constraints = []
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    a, b = map(int, line.split())
                    if convert_to_zero_based:
                        constraints.append((a-1, b-1))
                    else:
                        constraints.append((a, b))
                except ValueError:
                    continue
        return constraints
    def parse_temporal_constraints(self, text):
        constraints = {}
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    job, time = map(int, line.split())
                    constraints[job-1] = time
                except ValueError:
                    continue
        return constraints
    def solve(self):
        try:
            job_durations = [int(x.strip()) for x in self.job_durations.get().split(',')]
            machine_capacities = [int(x.strip()) for x in self.machine_caps.get().split(',')]
            num_machines = int(self.num_machines.get())
            # Validate basic inputs
            if len(job_durations) != int(self.num_jobs.get()):
                raise ValueError("Number of job durations doesn't match number of jobs")
            if len(machine_capacities) != num_machines:
                raise ValueError("Number of machine capacities doesn't match number of machines")
            # Parse constraints
            precedence_constraints = self.parse_constraints(self.precedence.get("1.0", tk.END))
            resource_constraints = self.parse_constraints(self.resources.get("1.0", tk.END))
            temporal_constraints = self.parse_temporal_constraints(self.temporal.get("1.0", tk.END))
            # Call appropriate algorithm
            if self.algorithm == "Backtracking":
                schedule, makespan = test.backtracking(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            else:
                schedule, makespan = test.genetic_algorithm(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            if schedule:
                self.plot_schedule(schedule, makespan)
                self.result_label.config(text=f"Makespan: {makespan}", foreground=COLORS['success'])
            else:
                self.result_label.config(text="No valid schedule found", foreground=COLORS['error'])
                self.ax.clear()
                self.canvas.draw()
        except Exception as e:
            self.result_label.config(text=str(e), foreground=COLORS['error'])
            self.ax.clear()
            self.canvas.draw()
    def plot_schedule(self, schedule, makespan):
        self.ax.clear()
        colors = plt.cm.Set3(np.linspace(0, 1, len(schedule)))
        # --- New logic to stack overlapping jobs ---
        # Get machine list
        machines = set(machine for _, (_, _, machine) in schedule.items())
        # Build time slots for each machine
        time_slots = {}
        for job, (start, end, machine) in schedule.items():
            for t in range(start, end):
                if (machine, t) not in time_slots:
                    time_slots[(machine, t)] = []
                time_slots[(machine, t)].append(job)
        # Assign slot positions to jobs
        job_positions = {}
        for job, (start, end, machine) in schedule.items():
            concurrent = set()
            for t in range(start, end):
                concurrent.update(time_slots.get((machine, t), []))
            used = {job_positions.get(j, -1) for j in concurrent if j in job_positions}
            pos = 0
            while pos in used:
                pos += 1
            job_positions[job] = pos
        # Plot each job in its slot
        yticks = []
        yticklabels = []
        for job, (start, end, machine) in schedule.items():
            position = job_positions[job]
            y_pos = f"Machine {machine+1} (Slot {position+1})"
            self.ax.barh(y=y_pos, width=end-start, left=start, color=colors[job], label=f"Job {job+1}")
            if y_pos not in yticklabels:
                yticks.append(y_pos)
                yticklabels.append(y_pos)
        self.ax.set_xlabel('Time', color=COLORS['text'])
        self.ax.set_title(f'Job Schedule', color=COLORS['primary'])
        self.ax.grid(True, color=COLORS['accent'], linestyle='--', alpha=0.3)
        # Remove duplicate legend entries
        handles, labels = self.ax.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        self.ax.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
        self.ax.invert_yaxis()
        self.fig.tight_layout()
        self.fig.subplots_adjust(right=0.85)
        self.canvas.draw()

class App:
    def __init__(self, root):
        self.root = root
        self.show_welcome()
    def show_welcome(self):
        for widget in self.root.winfo_children():
            widget.destroy()
        WelcomePage(self.root, self.show_scheduler)
    def show_scheduler(self, algorithm):
        for widget in self.root.winfo_children():
            widget.destroy()
        SchedulerPage(self.root, algorithm, self.show_welcome)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("Job Scheduler")
    root.geometry("900x700")
    app = App(root)
    root.mainloop() 
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
import test

# Modern color palette
COLORS = {
    'bg': '#F4F6FB',
    'primary': '#2563eb',
    'accent': '#22d3ee',
    'text': '#22223b',
    'card': '#fff',
    'button': '#2563eb',
    'button_hover': '#1e40af',
    'error': '#ef4444',
    'success': '#22c55e',
}

FONT = ('Segoe UI', 11)
TITLE_FONT = ('Segoe UI', 28, 'bold')
SUBTITLE_FONT = ('Segoe UI', 16)

class ScrollableFrame(ttk.Frame):
    def __init__(self, container, *args, **kwargs):
        super().__init__(container, *args, **kwargs)
        self.grid_rowconfigure(0, weight=1)
        self.grid_columnconfigure(0, weight=1)
        canvas = tk.Canvas(self, bg=COLORS['bg'], highlightthickness=0)
        scrollbar = ttk.Scrollbar(self, orient="vertical", command=canvas.yview)
        self.scrollable_frame = ttk.Frame(canvas, style='Card.TFrame')
        self.scrollable_frame.bind(
            "<Configure>",
            lambda e: canvas.configure(
                scrollregion=canvas.bbox("all")
            )
        )
        self.window_id = canvas.create_window((0, 0), window=self.scrollable_frame, anchor="nw")
        canvas.configure(yscrollcommand=scrollbar.set)
        canvas.grid(row=0, column=0, sticky="nsew")
        scrollbar.grid(row=0, column=1, sticky="ns")
        self.canvas = canvas
        self.bind_mousewheel(canvas)
        # Ensure the scrollable_frame always matches the canvas width
        canvas.bind(
            "<Configure>",
            lambda e: canvas.itemconfig(self.window_id, width=canvas.winfo_width())
        )
    def bind_mousewheel(self, widget):
        widget.bind_all("<MouseWheel>", self._on_mousewheel)
    def _on_mousewheel(self, event):
        self.canvas.yview_scroll(int(-1*(event.delta/120)), "units")

class ModernButton(ttk.Button):
    def __init__(self, master=None, **kwargs):
        super().__init__(master, **kwargs)
        self.default_style = kwargs.get('style', 'Modern.TButton')
        self.hover_style = 'Modern.TButton.Hover'
        self.bind('<Enter>', self.on_enter)
        self.bind('<Leave>', self.on_leave)
    def on_enter(self, e):
        self.configure(style=self.hover_style)
    def on_leave(self, e):
        self.configure(style=self.default_style)

class WelcomePage:
    def __init__(self, root, switch_callback):
        self.root = root
        self.switch_callback = switch_callback
        self.root.configure(bg=COLORS['bg'])
        style = ttk.Style()
        style.configure('Card.TFrame', background=COLORS['card'], relief='flat')
        style.configure('Title.TLabel', background=COLORS['card'], font=TITLE_FONT, foreground=COLORS['primary'])
        style.configure('Subtitle.TLabel', background=COLORS['card'], font=SUBTITLE_FONT, foreground=COLORS['text'])
        style.configure('Modern.TButton', font=FONT, background=COLORS['button'], foreground='white', padding=16)
        style.configure('Modern.TButton.Hover', background=COLORS['button_hover'], foreground='white')
        style.configure('Modern.TCheckbutton',
                       background=COLORS['card'],
                       foreground=COLORS['primary'],
                       font=FONT)
        # Centered card
        card = ttk.Frame(self.root, style='Card.TFrame', padding=40)
        card.place(relx=0.5, rely=0.5, anchor='center')
        ttk.Label(card, text='Job Scheduler', style='Title.TLabel').pack(pady=(0, 10))
        ttk.Label(card, text='Optimize your workflow', style='Subtitle.TLabel').pack(pady=(0, 30))
        ModernButton(card, text='Backtracking', command=lambda: self.switch_callback('Backtracking')).pack(fill='x', pady=8)
        ModernButton(card, text='Genetic Algorithm', command=lambda: self.switch_callback('Genetic')).pack(fill='x', pady=8)
        ttk.Label(card, text='by Your Name', style='Subtitle.TLabel', font=('Segoe UI', 10), foreground=COLORS['accent']).pack(pady=(30,0))

class SchedulerPage:
    def __init__(self, root, algorithm, back_callback):
        self.root = root
        self.algorithm = algorithm
        self.back_callback = back_callback
        self.root.configure(bg=COLORS['bg'])
        style = ttk.Style()
        style.configure('Card.TFrame', background=COLORS['card'], relief='flat')
        style.configure('Section.TLabelframe', background=COLORS['card'], font=FONT)
        style.configure('Section.TLabelframe.Label', background=COLORS['card'], font=FONT, foreground=COLORS['primary'])
        style.configure('Modern.TButton', font=FONT, background=COLORS['button'], foreground='white', padding=12)
        style.configure('Modern.TButton.Hover', background=COLORS['button_hover'], foreground='white')
        style.configure('Modern.TEntry', fieldbackground=COLORS['card'], borderwidth=1, relief='solid')
        style.configure('Modern.TCheckbutton',
                       background=COLORS['card'],
                       foreground=COLORS['primary'],
                       font=FONT)
        # Header
        header = ttk.Frame(self.root, style='Card.TFrame', padding=(30, 20))
        header.pack(fill='x', pady=(0, 10))
        ttk.Label(header, text='Configure Your Job Scheduling Problem', style='Title.TLabel').pack(side='left')
        ModernButton(header, text='Back', command=self.back_callback, style='Modern.TButton').pack(side='right')
        # Main content (now scrollable)
        scrollable = ScrollableFrame(self.root)
        scrollable.pack(fill='both', expand=True)
        content = scrollable.scrollable_frame
        content.grid_columnconfigure(0, weight=1)
        # Input section
        input_card = ttk.Labelframe(content, text='Basic Data', style='Section.TLabelframe', padding=20)
        input_card.grid(row=0, column=0, sticky='ew', padx=10, pady=10)
        input_card.grid_columnconfigure(1, weight=1)
        ttk.Label(input_card, text='Number of Jobs:', style='Subtitle.TLabel').grid(row=0, column=0, sticky='w', pady=4)
        self.num_jobs = ttk.Entry(input_card, width=10, style='Modern.TEntry')
        self.num_jobs.grid(row=0, column=1, pady=4, sticky='ew')
        self.num_jobs.insert(0, '2')
        ttk.Label(input_card, text='Job Durations (comma-separated):', style='Subtitle.TLabel').grid(row=1, column=0, sticky='w', pady=4)
        self.job_durations = ttk.Entry(input_card, width=30, style='Modern.TEntry')
        self.job_durations.grid(row=1, column=1, pady=4, sticky='ew')
        self.job_durations.insert(0, '2,3')
        ttk.Label(input_card, text='Number of Machines:', style='Subtitle.TLabel').grid(row=2, column=0, sticky='w', pady=4)
        self.num_machines = ttk.Entry(input_card, width=10, style='Modern.TEntry')
        self.num_machines.grid(row=2, column=1, pady=4, sticky='ew')
        self.num_machines.insert(0, '2')
        ttk.Label(input_card, text='Machine Capacities (comma-separated):', style='Subtitle.TLabel').grid(row=3, column=0, sticky='w', pady=4)
        self.machine_caps = ttk.Entry(input_card, width=30, style='Modern.TEntry')
        self.machine_caps.grid(row=3, column=1, pady=4, sticky='ew')
        self.machine_caps.insert(0, '1,1')
        # Algorithm selection (show as label, but keep for logic)
        ttk.Label(input_card, text='Algorithm:', style='Subtitle.TLabel').grid(row=4, column=0, sticky='w', pady=4)
        self.algorithm_label = ttk.Label(input_card, text=algorithm, style='Subtitle.TLabel', foreground=COLORS['primary'])
        self.algorithm_label.grid(row=4, column=1, pady=4, sticky='w')
        # Auto-assign option
        self.auto_split = tk.BooleanVar(value=True)
        ttk.Checkbutton(input_card, text="Auto-assign jobs to machines", variable=self.auto_split, style='Modern.TCheckbutton').grid(row=5, column=0, columnspan=2, sticky="w", pady=4)
        # Constraints section
        constraints_card = ttk.Labelframe(content, text='Constraints', style='Section.TLabelframe', padding=20)
        constraints_card.grid(row=1, column=0, sticky='ew', padx=10, pady=10)
        ttk.Label(constraints_card, text='Precedence Constraints (job1 job2 per line):', style='Subtitle.TLabel').pack(anchor='w', pady=(0,2))
        self.precedence = scrolledtext.ScrolledText(constraints_card, width=30, height=2, font=FONT)
        self.precedence.pack(fill='x', pady=2)
        ttk.Label(constraints_card, text='Resource Constraints (job machine per line):', style='Subtitle.TLabel').pack(anchor='w', pady=(8,2))
        self.resources = scrolledtext.ScrolledText(constraints_card, width=30, height=2, font=FONT)
        self.resources.pack(fill='x', pady=2)
        ttk.Label(constraints_card, text='Temporal Constraints (job time per line):', style='Subtitle.TLabel').pack(anchor='w', pady=(8,2))
        self.temporal = scrolledtext.ScrolledText(constraints_card, width=30, height=2, font=FONT)
        self.temporal.pack(fill='x', pady=2)
        # Solve button
        ModernButton(content, text='Solve', command=self.solve, style='Modern.TButton').grid(row=2, column=0, pady=20, sticky='ew')
        # Results area
        self.result_card = ttk.Labelframe(content, text='Schedule', style='Section.TLabelframe', padding=20)
        self.result_card.grid(row=3, column=0, sticky='ew', padx=10, pady=10)
        self.result_label = ttk.Label(self.result_card, text='', style='Subtitle.TLabel')
        self.result_label.pack()
        self.fig, self.ax = plt.subplots(figsize=(10, 3))
        self.canvas = FigureCanvasTkAgg(self.fig, master=self.result_card)
        self.canvas.get_tk_widget().pack(fill='both', expand=True)
        # Configure grid weights
        content.grid_rowconfigure(3, weight=1)
        content.grid_columnconfigure(0, weight=1)
    def parse_constraints(self, text, convert_to_zero_based=True):
        constraints = []
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    a, b = map(int, line.split())
                    if convert_to_zero_based:
                        constraints.append((a-1, b-1))
                    else:
                        constraints.append((a, b))
                except ValueError:
                    continue
        return constraints
    def parse_temporal_constraints(self, text):
        constraints = {}
        for line in text.split('\n'):
            line = line.strip()
            if line:
                try:
                    job, time = map(int, line.split())
                    constraints[job-1] = time
                except ValueError:
                    continue
        return constraints
    def solve(self):
        try:
            job_durations = [int(x.strip()) for x in self.job_durations.get().split(',')]
            machine_capacities = [int(x.strip()) for x in self.machine_caps.get().split(',')]
            num_machines = int(self.num_machines.get())
            # Validate basic inputs
            if len(job_durations) != int(self.num_jobs.get()):
                raise ValueError("Number of job durations doesn't match number of jobs")
            if len(machine_capacities) != num_machines:
                raise ValueError("Number of machine capacities doesn't match number of machines")
            # Parse constraints
            precedence_constraints = self.parse_constraints(self.precedence.get("1.0", tk.END))
            resource_constraints = self.parse_constraints(self.resources.get("1.0", tk.END))
            temporal_constraints = self.parse_temporal_constraints(self.temporal.get("1.0", tk.END))
            # Call appropriate algorithm
            if self.algorithm == "Backtracking":
                schedule, makespan = test.backtracking(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            else:
                schedule, makespan = test.genetic_algorithm(
                    job_durations, num_machines, precedence_constraints,
                    resource_constraints, machine_capacities, self.auto_split.get(), temporal_constraints
                )
            if schedule:
                self.plot_schedule(schedule, makespan)
                self.result_label.config(text=f"Makespan: {makespan}", foreground=COLORS['success'])
            else:
                self.result_label.config(text="No valid schedule found", foreground=COLORS['error'])
                self.ax.clear()
                self.canvas.draw()
        except Exception as e:
            self.result_label.config(text=str(e), foreground=COLORS['error'])
            self.ax.clear()
            self.canvas.draw()
    def plot_schedule(self, schedule, makespan):
        self.ax.clear()
        colors = plt.cm.Set3(np.linspace(0, 1, len(schedule)))
        # --- New logic to stack overlapping jobs ---
        # Get machine list
        machines = set(machine for _, (_, _, machine) in schedule.items())
        # Build time slots for each machine
        time_slots = {}
        for job, (start, end, machine) in schedule.items():
            for t in range(start, end):
                if (machine, t) not in time_slots:
                    time_slots[(machine, t)] = []
                time_slots[(machine, t)].append(job)
        # Assign slot positions to jobs
        job_positions = {}
        for job, (start, end, machine) in schedule.items():
            concurrent = set()
            for t in range(start, end):
                concurrent.update(time_slots.get((machine, t), []))
            used = {job_positions.get(j, -1) for j in concurrent if j in job_positions}
            pos = 0
            while pos in used:
                pos += 1
            job_positions[job] = pos
        # Plot each job in its slot
        yticks = []
        yticklabels = []
        for job, (start, end, machine) in schedule.items():
            position = job_positions[job]
            y_pos = f"Machine {machine+1} (Slot {position+1})"
            self.ax.barh(y=y_pos, width=end-start, left=start, color=colors[job], label=f"Job {job+1}")
            if y_pos not in yticklabels:
                yticks.append(y_pos)
                yticklabels.append(y_pos)
        self.ax.set_xlabel('Time', color=COLORS['text'])
        self.ax.set_title(f'Job Schedule', color=COLORS['primary'])
        self.ax.grid(True, color=COLORS['accent'], linestyle='--', alpha=0.3)
        # Remove duplicate legend entries
        handles, labels = self.ax.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        self.ax.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
        self.ax.invert_yaxis()
        self.fig.tight_layout()
        self.fig.subplots_adjust(right=0.85)
        self.canvas.draw()

class App:
    def __init__(self, root):
        self.root = root
        self.show_welcome()
    def show_welcome(self):
        for widget in self.root.winfo_children():
            widget.destroy()
        WelcomePage(self.root, self.show_scheduler)
    def show_scheduler(self, algorithm):
        for widget in self.root.winfo_children():
            widget.destroy()
        SchedulerPage(self.root, algorithm, self.show_welcome)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("Job Scheduler")
    root.geometry("900x700")
    app = App(root)
    root.mainloop() 
```
